```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = 'D:\Guvi\Capstone\Data Analyst Intern Cooking Assignment - Excel.xlsx'  

# Load each sheet into a separate DataFrame
user_details = pd.read_excel(file_path, sheet_name='UserDetails.csv')
cooking_sessions = pd.read_excel(file_path, sheet_name='CookingSessions.csv')
order_details = pd.read_excel(file_path, sheet_name='OrderDetails.csv')
```

```python
order_details["Rating"].fillna(order_details.Rating.mode()[0],inplace= True)
```

```python
# Check for duplicates and remove them
user_details.drop_duplicates(inplace=True)
cooking_sessions.drop_duplicates(inplace=True)
order_details.drop_duplicates(inplace=True)
```

```python
order_cooking_merged = pd.merge(order_details, cooking_sessions, on='Session ID', how='inner')
```

```python
order_cooking_merged.info()
```

```python
# Drop redundant columns if necessary (e.g., duplicate User ID or Dish Name columns)
order_cooking_merged = order_cooking_merged.drop(columns=['User ID_y', 'Dish Name_y', 'Meal Type_y'])

# Rename columns for clarity
order_cooking_merged.rename(columns={
    'User ID_x': 'User ID',
    'Dish Name_x': 'Dish Name',
    'Meal Type_x': 'Meal Type'
}, inplace=True)
```

```python
order_cooking_merged.head()
```

```python
# Merge the result with UserDetails on 'User ID'
final_merged_data = pd.merge(order_cooking_merged, user_details, on='User ID', how='inner')
```

```python
final_merged_data
```

```python

# Popular Dishes
popular_dishes = final_merged_data['Dish Name'].value_counts()
print(popular_dishes)
```

```python
# Popular Dishes Bar Plot
plt.figure(figsize=(10, 6))
popular_dishes.plot(kind='bar', color='skyblue')
plt.title('Most Ordered Dishes')
plt.xlabel('Dish Name')
plt.ylabel('Number of Orders')
plt.show()
```

```python
# Relationship between Cooking Sessions and Orders
session_vs_order = final_merged_data.groupby(['Dish Name'])[['Session Rating', 'Rating']].mean().reset_index()
print(session_vs_order)
```

```python
# Session Rating vs Order Rating
plt.figure(figsize=(10, 6))
sns.scatterplot(data=session_vs_order, x='Session Rating', y='Rating', hue='Dish Name', palette='viridis', s=100)
plt.title('Session Rating vs Order Rating')
plt.xlabel('Session Rating')
plt.ylabel('Order Rating')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
# Demographic Analysis
demographic_analysis = final_merged_data.groupby(['Age', 'Favorite Meal'])['Order ID'].count().unstack().fillna(0)
print(demographic_analysis)
```

```python
demographic_analysis_stacked = final_merged_data.groupby(['Age', 'Favorite Meal'])['Order ID'].count().unstack(fill_value=0)

demographic_analysis_stacked.plot(
    kind='bar',
    stacked=True,
    figsize=(12, 8),
    color=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728']
)
plt.title('Demographic Analysis: Orders by Age and Favorite Meal (Stacked Bar)', fontsize=14)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Number of Orders', fontsize=12)
plt.legend(title='Favorite Meal', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.violinplot(
    data=final_merged_data,
    x='Favorite Meal',
    y='Age',
    scale='count',
    inner='quartile',
    palette='pastel'
)
plt.title('Demographics Analysis: Age Distribution by Favorite Meal', fontsize=14)
plt.xlabel('Favorite Meal', fontsize=12)
plt.ylabel('Age', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
order_trend = final_merged_data.groupby('Order Date')['Order ID'].count().reset_index()

plt.figure(figsize=(12, 6))
sns.lineplot(data=order_trend, x='Order Date', y='Order ID', marker='o', color='teal')
plt.title('Trends in Total Orders Over Time', fontsize=14)
plt.xlabel('Order Date', fontsize=12)
plt.ylabel('Number of Orders', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
popular_dishes = final_merged_data.groupby(['Meal Type', 'Dish Name'])['Order ID'].count().reset_index()

plt.figure(figsize=(12, 6))
sns.barplot(
    data=popular_dishes,
    x='Dish Name',
    y='Order ID',
    hue='Meal Type',
    palette='Set2'
)
plt.title('Popularity of Dishes Across Meal Types', fontsize=14)
plt.xlabel('Dish Name', fontsize=12)
plt.ylabel('Number of Orders', fontsize=12)
plt.xticks(rotation=45)
plt.legend(title='Meal Type')
plt.tight_layout()
plt.show()
```

```python
favorite_meals = user_details['Favorite Meal'].value_counts()

plt.figure(figsize=(8, 8))
favorite_meals.plot(
    kind='pie', autopct='%1.1f%%', startangle=140, colors=['#ff9999','#66b3ff','#99ff99']
)
plt.title('Distribution of Favorite Meals', fontsize=14)
plt.ylabel('')
plt.show()
```
